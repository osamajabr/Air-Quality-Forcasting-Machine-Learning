# Air Quality Forcasting

```python
import pandas as pd
import numpy as np
```

```python
# Loading the data from csv file to pandas dataframe
air_quality_data = pd.read_csv('AirQualityUCI.csv', sep=';', decimal=',')

# Upon Observation:
# All values in the CSV file are delimited by semicolons (';').
#Several columns use commas as decimal points.
```

```python
air_quality_data.head()
```

```python
# Removing the last 2 columns from the dataframe
air_quality_data = air_quality_data.iloc[:, :-2]

# Rrinting the first 5 rows of the dataframe
air_quality_data.head()
```

```python
air_quality_data.tail()
```

```python
air_quality_data.shape
```

```python
# Removing rows that are entirely NaNs
air_quality_data = air_quality_data.dropna(how='all')

# Printing the cleaned DataFrame
air_quality_data.head()
```

```python
air_quality_data.shape
```

```python
air_quality_data.tail()
```

```python
# Getting some info about the data
air_quality_data.info()
```

```python
# Checking the number of missing values in the dataframe
air_quality_data.isnull().sum()
```

This shows that there are no missing values in the dataset. But the actual missing values are tagged with the value -200. This is described in UCI repository documentation.

```python
# Counting the number of times -200 appears in the data
air_quality_data.isin([-200]).sum(axis=0)
```

## Handling Missing Values

```python
#  Replacing all -200 values in the dataset with NaN to indicate missing data
air_quality_data = air_quality_data.replace(-200, np.nan)
```

```python
air_quality_data.isnull().sum()
```

```python
air_quality_data.mean()
```

```python
# Replacing the missing values with the mean value of each column
air_quality_data = air_quality_data.fillna(air_quality_data.mean())
```

```python
# Checking the number of missing values in the dataframe
air_quality_data.isnull().sum()
```

## Forcasting With FB Prophet Algorithm

```python
# Converting the date from DD/MM/YYYY to YYYY-MM-DD
date_info = pd.to_datetime(air_quality_data['Date'], dayfirst=True)

# Printing the converted date information
print(date_info)
```

```python
# Extracting the time information
time_info = air_quality_data['Time']

# Printing the time information
print(time_info)
```

```python
# Printing the data types of Variables
print(type(date_info))
print(type(time_info))
```

```python
# Combining 2 series to a pandas dataframe
date_time = pd.concat([date_info, time_info], axis=1)
date_time.head()
```

```python
date_time.shape
```

```python
# Combining date and time
date_time['ds'] = date_time['Date'].astype(str) + ' ' + date_time['Time'].astype(str)

date_time.head()
```

```python
date_time.info()
```

```python
# Replacing periods with colons in the 'ds' column to ensure proper time format
date_time['ds'] = date_time['ds'].str.replace('.', ':', regex=False)

# Converting 'ds' to datetime format
data = pd.DataFrame()
data['ds'] = pd.to_datetime(date_time['ds'])

data.head()
```

```python
# Assigning the 'RH' to the 'y' column
data['y'] = air_quality_data['RH']
data.head()
```

# Modelling

```python
# Installing the Prophet library via pip.
! pip install prophet
```

```python
from prophet import Prophet
```

```python
# Training the Prophet model
model = Prophet()
model.fit(data)
```

```python
# Creating future prediction dates for 365 hours
future = model. make_future_dataframe(periods=365, freq='H')
future.tail()
```

```python
# Predicting future values
forecast = model.predict (future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
# Plotting the forecast results
fig1 = model. plot (forecast)
```

```python
# Plotting forecast components.
fig2 = model. plot_components (forecast)
```
